**Tuple** - only diffrence from list is bracket and error. Tuple contains() bracket instead of square bracket [].

```python
#eg:- string
a='cnjfdweoldelmf' "ckedjnkdel" #which always comes with double or single quotes is called string
type(a)
```

### 4 Set
**A set is a collection which is unordered, unchangeable*, and unindexed.Sets are used to store multiple items in a single variable.**

```python
s={23,45,67,889}
print(type(s))
print(s)
```

```python
# for empty set
s=set()
print(type(s))
print(s)
```

```python
#keys are unique:
d={1:"one",2:"two",3:"three",1:"cbjsdd"} # it always shows the latest or last value of key
print(d)
```
